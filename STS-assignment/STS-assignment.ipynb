```python
# import necessary libraries
import warnings
warnings.filterwarnings("ignore")

import numpy as np
import pandas as pd

from sklearn.model_selection import train_test_split
from matplotlib import pyplot as plt
import seaborn as sns
from gensim.models import KeyedVectors

```

```python
#word2vec download link (Size ~ 1.5GB): https://drive.google.com/file/d/0B7XkCwpI5KDYNlNUTTlSS21pQmM/edit
path = '/gdrive/My Drive/Colab Notebooks/RNN/RNN Variants/GoogleNews-vectors-negative300.bin.gz'
model = KeyedVectors.load_word2vec_format(path, binary=True)
```

```python
data = pd.read_csv('/gdrive/My Drive/Colab Notebooks/STS-assignment/Precily_Text_Similarity.csv')
```

```python
result = []
#F.index.tolist()
for i in data.index.tolist():
  text1 = data.iloc[i][0]
  text2 = data.iloc[i][1]
  #calculate distance between two sentences using WMD algorithm

  distance = model.wmdistance(text1.lower(), text2.lower())
  result.append(distance)

result
```

```python
type(result)
```

```python
result = pd.DataFrame({'Similarity':result})
#result = pd.DataFrame(result)
```
